Observation: There are a lot of columns with missing values. In order to get a clear perspecctive.
    We write a function to output only missing values and percentage of missing values in each column where they occur

```python
def missing_values_table(df):   #source : https://www.analyticsvidhya.com/blog/2020/06/feature-engineering-guide-data-science-hackathons/
       mis_val = df.isnull().sum()
       mis_val_percent = 100 * df.isnull().sum() / len(df)
       mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
       mis_val_table_ren_columns = mis_val_table.rename(
       columns = {0 : 'Missing Values', 1 : '% of Total Values'})
       mis_val_table_ren_columns = mis_val_table_ren_columns[
           mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
       '% of Total Values', ascending=False).round(1)
       print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"     
           "There are " + str(mis_val_table_ren_columns.shape[0]) +
             " columns that have missing values.")
       return mis_val_table_ren_columns
    
missing_values_table(data) #calling the function
```

From the table above, we have a good visual representation of missing values as we can see the distribution of missing values.
We are going to drop columns that have its percentage of null values to be above 75%

```python
data=data[data.columns[data.isnull().mean() < 0.75]]
```

```python
 #check again for missing values
missing_values_table(data)
```

**1.Using RandomForestClassifier**

```python
X = data.drop(['date','failure'],axis=1).values
Y = data['failure'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
#Standardizing the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = RandomForestClassifier(n_jobs=-1)
estimators = np.arange(1, 33, 10)
scores = []
for n in estimators:
    model.set_params(n_estimators=n)
    model.fit(X_train, y_train)
    scores.append(model.score(X_test, y_test))
plt.title("Effect of n_estimators")
plt.xlabel("n_estimator")
plt.ylabel("score")
plt.plot(estimators, scores)
```

Plotting an elbow curve

```python
model.set_params(n_estimators=22)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

We have a  model score of 0.9939438306934647. This seems too high. We have to carry out cross validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
cv = KFold(n_splits=10); cv
```

```python

results = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')
```

```python
print(results)
```

```python
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```

Observation: Using RandomForestClassifier we got a score of 0.994 and this prompted us to crss validate to confirm there 
             is no overfitting. Using a 10 fold cross validation we have an average accuracy of 0.995
